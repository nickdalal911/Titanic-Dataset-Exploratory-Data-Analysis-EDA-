Exploratory Data Analysis on Titanic Dataset

```python
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
%matplotlib inline
```

```python
ship=sns.load_dataset('titanic')
```

```python
ship.head()
```

```python
ship.info()
```

```python
ship.isnull().sum()
```

```python
ship['age'].fillna(ship['age'].mean(),inplace=True)
```

```python
ship.isnull().sum()
```

```python
ship['embarked'].value_counts()
```

```python

ship['embarked'].fillna('S',inplace=True)
```

```python
ship.isnull().sum()
```

```python
ship.drop('deck',axis=1,inplace=True)
```

```python
ship.isnull().sum()
```

```python
ship['embark_town'].value_counts()
```

```python
ship['embark_town'].fillna('Southampton',inplace=True)
```

```python
ship.isnull().sum()
```

```python
ship
```

```python
sns.countplot(x='survived', data=ship)
plt.title("Survival Count")
```

```python
sns.countplot(x='pclass', data=ship)
plt.title("Passenger Class Distribution")
```

```python
sns.histplot(ship['age'], kde=True, bins=30)
plt.title("Age Distribution")
```

```python
sns.countplot(x='sex', data=ship)
plt.title("Gender Count")
```

```python
sns.countplot(x='embarked', data=ship)
plt.title("Embarked Port Distribution")
```

```python
sns.barplot(x='sex', y='survived', data=ship)
plt.title("Survival Rate by Gender")
```

```python
sns.barplot(x='pclass', y='survived', data=ship)
plt.title("Survival Rate by Class")
```

```python
sns.boxplot(x='survived', y='age', data=ship)
plt.title("Age Distribution by Survival")
```

```python
sns.violinplot(x='pclass', y='age', hue='survived', data=ship, split=True)
plt.title("Age & Survival by Class")
```

```python
sns.pairplot(ship[['survived', 'pclass', 'age', 'fare', 'sex']])
```

```python
ship.groupby('sex')['survived'].mean()
ship.groupby('pclass')['survived'].mean()
ship.pivot_table(index='sex', columns='pclass', values='survived')
```
